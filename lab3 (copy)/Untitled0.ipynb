```python
import csv
m= open ('cig.csv','rb')
reader = csv.reader(m)
index = 0

tar = []
tin = []

for row in reader:
    index += 1 
    n = row[0]
    n = n.replace("\t", " ")
    n = n.split()
    n = n[1:] 
    n = map(float,n)
    tar.append(n[0])
    tin.append(n[1])
m.close () 

from pylab import *
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('cigarettes')
ax.set_xlabel('tar')
ax.set_ylabel('nicotine')

scatter(tar,tin)
savefig("scatter_cig.png")
```
