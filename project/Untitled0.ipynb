```python
import csv
m = open ('data.csv','rb')
r = csv.reader (m)
rownum = 0
ta = []
da = [] 

for row in r:
    if rownum != 0:
        n = row[0]
        n = n.replace("\t", " ")
        n = n.split()
        n = n[0:] 
        n = map(float,n)
        ta.append(n[0])
        da.append(n[1])
    rownum +=1
print ta
print da

m.close () 
```

```python
from pylab import *
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('IQ compar to the brain size')
ax.set_xlabel('brain size')
ax.set_ylabel('IQ')

scatter(ta, da)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('IQ compar to the brain size')
ax.set_xlabel('brain size')
ax.set_ylabel('IQ')

plot(ta, da)
```
